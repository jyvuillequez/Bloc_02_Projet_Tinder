# Data preparation et cleaning

```python
# fonctions utiles
def missing_report(df: pd.DataFrame) -> pd.DataFrame:
    """Retourne un tableau tri des % de valeurs manquantes par colonne."""
    rep = (df.isna().mean() * 100).sort_values(ascending=False).to_frame("missing_%")
    rep["missing_n"] = df.isna().sum()
    rep["dtype"] = df.dtypes.astype(str)
    return rep

def safe_cols(df: pd.DataFrame, cols: list[str]) -> list[str]:
    """Ne garde que les colonnes rellement prsentes dans le df."""
    return [c for c in cols if c in df.columns]
```

```python
# Standardisation des noms de colonnes
df = df.copy()
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_", regex=False)
)
```

```python
# Gestion des duplications de donnes 
dup_count = df.duplicated().sum()
print("Duplicate rows:", dup_count)
if dup_count > 0:
    df = df.drop_duplicates()
    print("New shape after drop_duplicates:", df.shape)

# Rappel des valeurs manquantes (avant nettoyage)
miss_before = missing_report(df)
display(miss_before.head(30))
```

```python
# Target variable cleaning (decision / second date)
# Dans le dataset, la cible est souvent 'dec' (0/1 : veut revoir la personne)
# Nettoyage et forage du type
target_candidates = ["dec", "decision", "match", "like", "second_date"]
target = next((c for c in target_candidates if c in df.columns), None)

if target is None:
    raise ValueError("Aucune colonne cible trouve (dec/decision/match...). Le nom de ta colonne target est  vrifier.")

# On conserve uniquement les lignes o la target est connue
df = df[df[target].notna()].copy()

# Force en int 0/1 si possible
df[target] = df[target].astype(int)

print("Target distribution:")
display(df[target].value_counts(dropna=False))
print("Target rate (yes):", df[target].mean().round(3))
```

```python
# Catgorisation des variables (align avec lobjectif d'analyse)
# Notes donnes  l'autre (perception du partenaire)
rating_cols = [
    "attr", "sinc", "intel", "fun", "amb", "shar"  # noms classiques du dataset
]

# Variables dmographiques / contexte (selon dispo)
demo_cols = [
    "gender", "age", "race", "field_cd", "career_c", "goal", 
    "order", "round", "position", "wave"
]

# Variables auto-perception (souvent suffixes _o ou _s selon le dataset)
self_cols_candidates = [
    "attr_s", "sinc_s", "intel_s", "fun_s", "amb_s", "shar_s",
    "attr_o", "sinc_o", "intel_o", "fun_o", "amb_o", "shar_o"
]

# Variables "importance dclare" (souvent pref / importance)
pref_cols_candidates = [
    "pref_o_attractive", "pref_o_sincere", "pref_o_intelligence", "pref_o_funny",
    "pref_o_ambitious", "pref_o_shared_interests",
    "attr1_1", "sinc1_1", "intel1_1", "fun1_1", "amb1_1", "shar1_1"
]

keep_cols = [target] + safe_cols(df, rating_cols) + safe_cols(df, demo_cols) \
            + safe_cols(df, self_cols_candidates) + safe_cols(df, pref_cols_candidates)

df_clean = df[keep_cols].copy()
print("Kept columns:", len(df_clean.columns))
display(df_clean.head(3))
```

```python
# Cleaning des ratings
rating_cols_present = safe_cols(df_clean, rating_cols)

for c in rating_cols_present:
    # Convert to numeric (si incohrence de type)
    df_clean[c] = pd.to_numeric(df_clean[c], errors="coerce")

    # Remove impossible values (ex: <0 or >10) en les mettant en NaN
    df_clean.loc[(df_clean[c] < 0) | (df_clean[c] > 10), c] = np.nan

# Dcision sur les NaN des ratings :
# - option A (simple & propre pour certif) : on garde les lignes compltes pour analyses rating-centric
df_ratings = df_clean.dropna(subset=rating_cols_present).copy()

print("Shape df_clean:", df_clean.shape)
print("Shape df_ratings (no missing ratings):", df_ratings.shape)
```

```python
# Cleaning des donnes dmographiques
if "age" in df_clean.columns:
    df_clean["age"] = pd.to_numeric(df_clean["age"], errors="coerce")
    df_clean.loc[(df_clean["age"] < 18) | (df_clean["age"] > 80), "age"] = np.nan

# Gender : dans ce dataset souvent 0=female 1=male
if "gender" in df_clean.columns:
    df_clean["gender"] = pd.to_numeric(df_clean["gender"], errors="coerce")
    # option : mapping lisible pour viz
    df_clean["gender_label"] = df_clean["gender"].map({0: "Female", 1: "Male"})

# Order / round : numeric
for c in ["order", "round", "position"]:
    if c in df_clean.columns:
        df_clean[c] = pd.to_numeric(df_clean[c], errors="coerce")
```

```python
# Valeurs manquantes (aprs nettoyage)
miss_after = missing_report(df_clean)
display(miss_after.head(30))
```

```python
# Sauvegarde du dataset nettoy
df_clean.to_csv("../data/outputs/speed_dating_clean.csv", index=False)
df_ratings.to_csv("../data/outputs/speed_dating_clean_ratings_only.csv", index=False)

print("Cleaning OK.")
```
